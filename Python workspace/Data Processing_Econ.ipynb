```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/nikmag/Desktop/Visualisation/project-thehighersociety/src/assets/SYB61_T14_Gross Value Added by Economic Activity.csv")
```

```python
df.drop(columns=['Region/Country/Area', 'Footnotes', 'Source'], axis=1, inplace=True)
```

```python
df.rename(columns={'Unnamed: 1': 'Country'}, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.Series.value_counts()
```

```python
df.head()
```

```python
pivot = df.pivot(columns='Series', values='Value')
pivot.reset_index(inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
a = pd.merge(df, pivot, on = 'index')
```

```python
a
```

```python
b = a.groupby(as_index=False ,by=['Country', 'Year']).agg({'Agriculture, hunting, forestry and fishing (% of gross value added)':'first', 'Industry (% of gross value added)': 'first', 'Services (% of gross value added)': 'first'})
```

```python
b.fillna(np.mean, inplace=True)
```

```python
b.isnull().sum()
```

```python
b.to_csv("processed_economic_activity.csv", index=False)
```
