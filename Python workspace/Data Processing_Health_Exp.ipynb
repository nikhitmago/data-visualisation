```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/Users/nikmag/Desktop/Visualisation/project-thehighersociety/src/assets/SYB61_T11_Expenditure on Health.csv")
```

```python
df.drop(columns=['Region/Country/Area', 'Footnotes', 'Source'], axis=1, inplace=True)
```

```python
df.rename(columns={'Unnamed: 1': 'Country'}, inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.Series.value_counts()
```

```python
df
```

```python
pivot = df.pivot(columns='Series', values='Value')
pivot.reset_index(inplace=True)
```

```python
df.reset_index(inplace=True)
```

```python
a = pd.merge(df, pivot, on = 'index')
```

```python
b = a.groupby(as_index=False ,by=['Country', 'Year']).agg({'Current health expenditure (% of GDP)':'first', 'Domestic general government health expenditure (% of total government expenditure)': 'first'})
```

```python
b.fillna(np.mean, inplace=True)
```

```python
b.isnull().sum()
```

```python
b.to_csv("processed_health.csv", index=False)
```
